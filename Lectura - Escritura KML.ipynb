```python
import geopandas as gpd
import fiona
import numpy as np
import pandas as pd
```

# Leer KML

```python
#gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'

#df = gpd.read_file(path_to_data+transac_file+'.kml', driver='KML')


gpd.io.file.fiona.drvsupport.supported_drivers['LIBKML'] = 'rw'


#fp=path_to_data+transac_file+'.kml'
fp='temp.kml'

gdf_list = []
for layer in fiona.listlayers(fp):    
    gdf = gpd.read_file(fp, driver='LIBKML', layer=layer)
    gdf_list.append(gdf)

gdf = gpd.GeoDataFrame(pd.concat(gdf_list, ignore_index=True))

gdf = gdf.set_crs(epsg=4326)

gdf
```

```python
#agregarle el CSR

gdf.to_file('temp.shp')
```

# Creando Geopandas



```python
df = pd.read_csv('Salida3.csv')
df.head()
```

```python
df.rename(columns = {'nombre':'Name'}, inplace = True)
df.rename(columns = {'captura':'timestamp'}, inplace = True)

# All DataFrame replace empty string
df = df.replace(np.nan, '', regex=True)
df['descripcion'][(df['descripcion']=='desc')]=''
df.head()
```

```python
df['description']=df['formacion']+'\n'+df['descripcion']+'\n<img src="'+df['ruta']+'" style="width:600px;height:337px;">'
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.lon, df.lat), crs="EPSG:4326"
)
#gdf
```

# Generamos KML

```python
fiona.supported_drivers['KML'] = 'rw'

gdf.to_file('test.kml', driver='KML')
```
